## Divisão de treino e teste

```python
df_treino,df_teste = train_test_split(dataset,test_size=0.3,random_state=45)
print(len(df_treino))
print(len(df_teste))
```

```python
x_treino = df_treino[train_features].copy()
y_treino = df_treino[target].copy()

x_teste = df_teste[train_features].copy()
y_teste = df_teste[target].copy()
```

## Modelo XGBOOST

### Otimização dos Parâmetros

```python
params = {
        'learning_rate':(0.01,1.0),
        'gamma':(0,10),
        'max_depth':(2,500),
        'min_child_weight':(1,300),
        'max_delta_step':(1,300),
        'subsample':(0,1),
        'colsample':(0,1),
        'n_estimators':(10,100)
}
```

```python
def xgboost_hyper_params(learning_rate,gamma,min_child_weight,max_delta_step,subsample,colsample,max_depth,n_estimators):
    max_depth = int(max_depth)
    n_estimators = int(n_estimators)
    clf = XGBClassifier(learning_rate = learning_rate,
                        gamma = gamma,
                        min_child_weight = min_child_weight,
                        max_delta_step = max_delta_step,
                        max_depth = max_depth,
                        n_estimators = n_estimators,
                        eval_metric = 'auc')
    
    return np.mean(cross_val_score(clf,x_treino,y_treino,cv=10,scoring='roc_auc'))
    
```

```python
otimizacao = BayesianOptimization(f=xgboost_hyper_params,pbounds = params,random_state=1)
otimizacao.maximize(n_iter=15, init_points=8, acq='ei')
parametros = otimizacao.max['params']
```

```python
parametros
```

### Treinamento do Modelo

```python
xgb_modelo = XGBClassifier(
                        learning_rate = 0.149,
                        gamma = 5.587,
                        min_child_weight = 290.5,
                        max_delta_step = 60.23,
                        max_depth = int(400.8),
                        n_estimators = int(38.21),
                        eval_metric = 'auc')
xgb_modelo.fit(x_treino,y_treino)
```

```python
def evaluate_model(df_train,df_test,target,model,train_features):

    df_train['Probability'] = model.predict_proba(df_train[train_features])[:,1]
    df_test['Probability'] = model.predict_proba(df_test[train_features])[:,1]

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (14, 6))
    # Plot AUC Curve

    fpr_train, tpr_train, threshold_train = roc_curve(df_train[target], df_train['Probability'])
    roc_auc_train = auc(fpr_train, tpr_train)

    fpr_test, tpr_test, threshold_test = roc_curve(df_test[target], df_test['Probability'])
    roc_auc_test = auc(fpr_test, tpr_test)
    sns.set(font_scale=1.5) 
    title = 'Receiver Operating Characteristic (ROC) Curve' 
    ax = axes[0]
    ax.plot(fpr_train , tpr_train , color='darkorange', label = 'AUC = {}'.format(round(roc_auc_train,3))) 
    ax.plot([0, 1], [0, 1], color='navy', linestyle='--')
    ax.legend(loc="lower right")
    ax.set(xlabel='False Positive Rate', ylabel='True Positive Rate')
    ax.set_title('Train')

    ax = axes[1]
    ax.plot(fpr_test , tpr_test , color='darkorange', label = 'AUC = {}'.format(round(roc_auc_test,3))) 
    ax.plot([0, 1], [0, 1], color='navy', linestyle='--')
    ax.legend(loc="lower right")
    ax.set(xlabel='False Positive Rate', ylabel='True Positive Rate')
    ax.set_title('Validation')


    fig.suptitle(title)
```

```python
evaluate_model(df_train = df_treino
               ,df_test = df_teste
               ,target  =  target
               ,model   = xgb_modelo
               ,train_features =  train_features)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
def plot_matriz_confusao(y_true_teste, y_pred_teste, group_names=None,
                         categories='auto', count=True, cbar=True,
                         xyticks=True, sum_stats=True, figsize=None,
                         cmap='viridis', title=None):

    cf = confusion_matrix(y_true_teste, y_pred_teste)

    blanks = ['' for i in range(cf.size)]

    if group_names and len(group_names) == cf.size:
        group_labels = ["{}\n".format(value) for value in group_names]
    else:
        group_labels = blanks

    if count:
        group_counts = ["{0:0.0f}\n".format(value) for value in cf.flatten()]
    else:
        group_counts = blanks

    box_labels = [f"{v1}{v2}".strip()
                  for v1, v2 in zip(group_labels, group_counts)]
    box_labels = np.asarray(box_labels).reshape(cf.shape[0], cf.shape[1])

    if sum_stats:

        accuracy = accuracy_score(y_true_teste, y_pred_teste)
        precision = precision_score(y_true_teste, y_pred_teste)
        recall = recall_score(y_true_teste, y_pred_teste)
        f1_score_metric = f1_score(y_true_teste, y_pred_teste)

        stats_text = "\n\nAcurácia={:0.3f}\nPrecisão={:0.3f}\nRecall={:0.3f}\nF1 Score={:0.3f}".format(
            accuracy, precision, recall, f1_score_metric)
    else:
        stats_text = ""

    if figsize == None:
        figsize = plt.rcParams.get('figure.figsize')

    if xyticks == False:
        categories = False

    plt.figure(figsize=figsize)
    sns.set(font_scale=1.4)  # for label size
    sns.heatmap(cf, annot=box_labels, fmt="", cmap=cmap, cbar=cbar,
                xticklabels=categories, yticklabels=categories)
    plt.ylabel('Valores verdadeiros', fontsize=16)
    plt.xlabel('Valores preditos' + stats_text, fontsize=16)

    if title:
        plt.title(title, fontsize=20)
```

```python
y_pred_dt = xgb_modelo.predict(x_teste)
```

```python
labels = ["True Neg","False Pos","False Neg","True Pos"]
categories = ["Não Fraude", "Fraude"]
plot_matriz_confusao(y_teste,
                      y_pred_dt,
                      group_names=labels,
                      categories=categories,
                      figsize=(8, 6), 
                      title="Matriz de confusão para o classificador")
```
